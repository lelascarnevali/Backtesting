```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Baixa os dados da ação desejada
df_weekly = yf.download("USDBRL=X", interval='1wk')
```

```python
# Calcula os retornos diários
df_weekly['Return'] = df_weekly['Close'].pct_change()

# Calcula o valor acumulado
df_weekly['Cumulative'] = (1 + df_weekly['Return']).cumprod()

# Calcula o drawdown
df_weekly['Peak'] = df_weekly['Cumulative'].cummax()
df_weekly['Drawdown'] = (df_weekly['Cumulative'] / df_weekly['Peak']) - 1
```

```python
plt.plot(df_weekly['Drawdown'])
plt.xlabel('Data')
plt.ylabel('Drawdown')
plt.title('Drawdown do Dólar')
plt.grid(True)
plt.show()
```
